```python
def checkHeliXExists(pdb_lines) :
    for line in pdb_lines:
        if(line[0:5]== 'HELIX'):
            return 1 ; 
    return 0 ; 
        
```

```python
representative_sequence = [];
flag = 0; 
for line in clusted_sorted_lines : 
    if(flag and line[1:8] != 'Cluster'):
        pdb_id = line.split('|')[0].split('>')[-1].split('_')[0];
        pdb_f = open(all_pdb_dir + '/'+(pdb_id).lower() +'.pdb');
        pdb_lines = pdb_f.readlines()
        if(checkHeliXExists(pdb_lines)):
            representative_sequence.append(pdb_id.lower());
            flag =0 ;
        else :
            continue;
    if(line[1:8] == 'Cluster'):
        flag = 1; 

```
